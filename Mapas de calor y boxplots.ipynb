```python
# Celda 1: importar y cargar
%matplotlib inline
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")

# Cargar dataset (ajusta ruta si hace falta)
df = pd.read_csv("diabetes.csv")

# Normalizar nombres de columnas a minúsculas (evita errores)
df.columns = df.columns.str.strip().str.lower()

# Vista rápida
print("Columnas:", list(df.columns))
df.head()
```

```python
# Celda 2: info y estadísticos
print("Shape:", df.shape)
display(df.info())
display(df.describe().round(2))
```

```python
# Celda 3: media, mediana, std, min, max
numericas = df.select_dtypes(include=[np.number]).columns.tolist()
stats = pd.DataFrame({
    'mean': df[numericas].mean(),
    'median': df[numericas].median(),
    'std': df[numericas].std(),
    'min': df[numericas].min(),
    'max': df[numericas].max()
})
display(stats.round(2))
```

```python
# Celda 4: ceros en columnas que suelen no tomar 0 y missing
check_zero_cols = ['glucose','bloodpressure','skinthickness','insulin','bmi']
present = [c for c in check_zero_cols if c in df.columns]

zeros_count = {c: int((df[c] == 0).sum()) for c in present}
na_count = df.isna().sum().to_dict()

print("Conteo de ceros (columnas relevantes):", zeros_count)
print("\nConteo de NaN por columna:")
display(pd.Series(na_count).sort_values(ascending=False))
```

```python
# Celda 5: barras para outcome (si existe)
if 'outcome' in df.columns:
    plt.figure(figsize=(5,4))
    sns.countplot(data=df, x='outcome')
    plt.title("Distribución de Outcome (0 = no, 1 = sí)")
    plt.xlabel("Outcome")
    plt.ylabel("Conteo")
    plt.tight_layout()
    plt.show()
else:
    print("No existe columna 'outcome' en este dataset.")
```

```python
# Celda 6: histogramas por variable numérica
for col in numericas:
    plt.figure(figsize=(6,3.5))
    sns.histplot(df[col].dropna(), bins=20, kde=True)
    plt.title(f'Histograma de {col}')
    plt.xlabel(col)
    plt.ylabel('Frecuencia')
    plt.tight_layout()
    plt.show()
```

```python
# Celda 7: boxplots
for col in numericas:
    plt.figure(figsize=(6,2.8))
    sns.boxplot(x=df[col].dropna())
    plt.title(f'Boxplot de {col}')
    plt.tight_layout()
    plt.show()
```

```python
# Celda 8: correlación y heatmap
corr = df[numericas].corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", vmin=-1, vmax=1, linewidths=0.4)
plt.title("Mapa de calor de correlaciones")
plt.tight_layout()
plt.show()

# Mostrar las correlaciones ordenadas con outcome (si existe)
if 'outcome' in corr.columns:
    print("\nCorrelación con Outcome (ordenadas):")
    display(corr['outcome'].sort_values(ascending=False).round(2))
```

```python
# Celda 9: outliers por IQR (resumen)
outlier_summary = []
for col in numericas:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    n_out = int(df[(df[col] < lower) | (df[col] > upper)].shape[0])
    outlier_summary.append((col, float(Q1), float(Q3), float(lower), float(upper), n_out))

outlier_df = pd.DataFrame(outlier_summary, columns=['variable','Q1','Q3','lower','upper','n_outliers'])
display(outlier_df.sort_values(by='n_outliers', ascending=False).round(2))
```

```python
# Celda 10: min-max por variable
ranges = df[numericas].agg(['min','max']).T
ranges.columns = ['min','max']
display(ranges.round(2))
```
